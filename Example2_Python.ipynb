## Histrograms

```python
# plot histrogram of gene expression (blue)
gene_expr.T.hist(figsize=(24,2), layout=(1,8), color="blue");
# plot histrogram of raw counts (high sequencing depth)
raw_count_high.T.hist(figsize=(24,2), layout=(1,8), color="red");
# plot histrogram of raw counts (mediun sequencing depth)
raw_count_med.T.hist(figsize=(24,2), layout=(1,8), color="orange");
# plot histrogram of raw counts (low sequencing depth)
raw_count_low.T.hist(figsize=(24,2), layout=(1,8), color="yellow");
```

## Sparsity

```python
# sparsity of simulated gene expression level
100 * sum((gene_expr == 0).sum()) / gene_expr.size
```

```python
# sparsity of simulated raw count with high sequencing depth
100 * sum((raw_count_high == 0).sum()) / raw_count_high.size
```

```python
# sparsity of simulated raw count with medium sequencing depth
100 * sum((raw_count_med == 0).sum()) / raw_count_med.size
```

```python
# sparsity of simulated raw count with low sequencing depth
100 * sum((raw_count_low == 0).sum()) / raw_count_low.size
```
